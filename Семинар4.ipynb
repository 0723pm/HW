***Задача 1***

```python
from pandas.core.describe import describe_categorical_1d
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

data = pd.read_csv('kc-house-data.csv', encoding='Windows-1251')
data.head()
```

**1.2. Изучите стоимости недвижимости**

```python
plt.figure(figsize = (6,4))
plt.hist(data['price'], bins = 50)
plt.title('Стоимость недвижимости');
plt.xlabel('Стоимость')
plt.ylabel('Количество')
```

##Вывод
Наибольшее количество недвижимости продается по цене 400 000-500 000.
Наименьшее количество недвижимости продается по цене выше 2 000 000.

**1.3. Изучите распределение жилой квадратуры**

```python
plt.figure(figsize = (6,4))
sns.histplot(data['sqft_living'], bins = 50)
plt.xlabel('Квадратные метры')
plt.ylabel('Количество')
plt.title('Распределение жилой квадратуры');
```

##Вывод
Максимальное количество недвижимости приобретается размером от 1700 кв. до 2100 кв.м.

**1.4. Изучите распределение года постройки**

```python
plt.figure(figsize = (6,4))
sns.histplot(data['yr_built'], bins = 50)
plt.xlabel('Год постройки')
plt.ylabel('Количество')
plt.title('Распределение годов постройки');
```

##Вывод
Наименьшее количество домов было построено в 1934-1938 годы.

Наибольшее количество домов было построено в 1775 год и 2009 год.

В среднем, колебания между большим и меньшим количеством построенных домов приходятся на каждые 7-9 лет.

**2.1. Изучите распределение домов от наличия вида на набережную**

```python
data_view = data['waterfront'].value_counts()
value = data_view.values
plt.pie(value,labels=['С видом на набережную', 'Без вида на набережную'],explode=[0,0.3],
        radius = 1.5, startangle=60,
        labeldistance=1.2, autopct='%.1f%%');
```

##Вывод

С видом на набережную продается 0,8% домов.
Без вида на набережную продается 99.2% домов.

***2.2. Изучите распределение этажей домов***

```python
data.floors.max(), data.floors.min()
data2 = data['floors'].value_counts(normalize=True)*100
data2.head()
```

```python
floors = data['floors'].value_counts()
floor_value = floors.values
floor_ind = floors.index
plt.figure(figsize=(12, 6))
plt.bar(floor_ind,floor_value, width=0.4, facecolor='c', alpha=0.75)
plt.title('Этажность')
plt.xlabel('Этажи')
plt.ylabel('Количество домов')
plt.grid(True)
plt.show()
```

##Вывод

***2.3. Изучите распределение состояния домов***

```python
data_condition = data['condition'].value_counts(sort=False)
data_condition
condition_val= data_condition.values
condition_ind = data_condition.index
condition_explode = [0,0,0,0.6,0.3]
colors = ["#FF69B4", "#98FB98", "#008080","#F0E68C", "#4169E1"]
plt.title("Состояние домов")
plt.pie(condition_val, labels=condition_ind, radius = 1.5,
        startangle=-20, colors=colors,
        rotatelabels=True, explode = condition_explode,
        labeldistance=1.1, autopct='%.1f%%')
plt.title("Состояние домов", bbox={'facecolor':'0.8', 'pad':4})
plt.show;
```

##Вывод
64.9% домов в хорошем состоянииЮ, что является 2/3 от общего количества домов.
34.2% домов в среднем состоянии, а также менее 1% домов в плохом состоянии.
